```python
quaterly_tourism=pd.read_csv("/content/Quarterly tourists in India - Q1.csv")
```

```python
quaterly_tourism.head()
```

```python
quaterly_tourism.columns
```

```python
quaterly_tourism['Year']
```

```python
#total earn from foreign in a decade
total_earn=quaterly_tourism['Foreign Exchange Earnings'].sum()
total_earn
```

```python
#total number of person over a decade
total_person=quaterly_tourism['Foreign Tourist Arrivals in India'].sum()
total_person
```

```python
#total earn on 1 person over a decade in percentage
per_person_earning=(total_earn/total_person)*100
per_person_earning
```

```python
#total earn on 1 person every year in percentage
quaterly_tourism['earn from per person']=(quaterly_tourism['Foreign Exchange Earnings']/quaterly_tourism['Foreign Tourist Arrivals in India'])*100
```
